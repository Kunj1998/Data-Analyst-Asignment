```python
# User Demographics Visualization
def user_demographics(country):
    filtered_df = df[df['Country'] == country]
    plt.figure(figsize=(10, 6))
    sns.countplot(data=filtered_df, x='Gender')
    plt.title(f'Number of Users by Gender in {country}')
    plt.xlabel('Gender')
    plt.ylabel('Number of Users')
    plt.show()
```

```python
# Subscription Overview
def subscription_overview(plan_duration):
    filtered_df = df[df['Plan Duration'] == plan_duration]
    plt.figure(figsize=(8, 8))
    filtered_df['Subscription Type'].value_counts().plot(kind='pie', autopct='%1.1f%%',explode=p,shadow=True)
    plt.title(f'Subscription Proportion for {plan_duration} Plan Duration')
    plt.ylabel('')
    plt.show()
```

```python
# Monthly Revenue Trend
def monthly_revenue_trend(segmented):
    plt.figure(figsize=(12, 6))
    if segmented:
        sns.lineplot(data=df, x='Join Date', y='Monthly Revenue',hue='Subscription Type' )
        plt.title('Total Monthly Revenue Trend Segmented by Subscription Type')
    else:
        df.groupby(df['Join Date'].dt.to_period('M'))['Monthly Revenue'].sum().plot()
        plt.title('Total Monthly Revenue Trend')
    plt.xlabel('Date')
    plt.ylabel('Total Revenue')
    plt.xticks(rotation=45)
    plt.show()
```

```python
# Interactive Dashboard
# User Demographics Visualization
country_dropdown = widgets.Dropdown(
    options=df['Country'].unique(),
    value=df['Country'].unique()[0],
    description='Country:'
)

interact(user_demographics, country=country_dropdown)
```

```python
# Create a dropdown widget for selecting the country
country_dropdown = widgets.Dropdown(options=['All'] + df['Country'].unique().tolist(),value='All',  description='Country')
```

```python
# Device used by Male and Female on Age Number
# Define a function to update the bar plot based on the selected Device
def Device_plot(country):
    if country == 'All':
        data_to_plot = df
    else:
        data_to_plot = df[df['Country'] == country]
    
    sns.barplot(x='Gender', y='Age', data=data_to_plot, hue='Device')
   

# Call the Device_plot function whenever the dropdown value changes
widgets.interactive(Device_plot, country=country_dropdown)
```

```python
# Implement tooltips for additional information on hover.

import holoviews as hv
from bokeh.models import HoverTool, CustomJSHover
# User Demographics Visualization with tooltips
def User_demographics(country):
    filtered_df = df[df['Country'] == country]
    
    # Create the count plot using Seaborn
    plt.figure(figsize=(10, 6))
    sns.countplot(data=filtered_df, x='Gender')
    plt.title(f'Number of Users by Gender in {country}')
    plt.xlabel('Gender')
    plt.ylabel('Number of Users')
    
    # Show the plot
    plt.show()

# Test the function
User_demographics('United States')

# Prepare the data
data = df[df['Country'] == 'United States']  # Example: Filter data for the United States
data_counts = data['Gender'].value_counts().reset_index()
data_counts.columns = ['Gender', 'Count']

# Create the HoverTool with tooltips
hover = HoverTool(tooltips=[("Gender", "@Gender"), ("Count", "@Count")])

# Create the scatter plot using Holoviews and Bokeh
scatter = hv.Scatter(data_counts, kdims='Gender', vdims=['Count']).opts(tools=[hover], size=10)

# Show the plot
scatter.redim.range(Count=(0, data_counts['Count'].max()))  # Set the range for Count
```
